```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
```

```python

X = 2 * np.random.rand(100, 1)  # 100 random points between [0, 2]
y = 4 + 3 * X + np.random.randn(100, 1)  # True relation: y = 4 + 3x + noise

plt.scatter(X, y)
plt.xlabel("X")
plt.ylabel("y")
plt.title("Synthetic Linear Data")
plt.show()
```

```python
def stochastic_gradient_descent(X, y, learning_rate=0.01, n_epochs=50):
    m = len(X)  # Number of training samples
    theta = np.random.randn(2, 1)  # Random initialization [θ₀, θ₁]
    X_b = np.c_[np.ones((m, 1)), X]  # Add x₀ = 1 for bias term (θ₀)

    cost_history = []

    for epoch in range(n_epochs):
        for i in range(m):  # Iterate over each sample
            random_idx = np.random.randint(m)  # Pick a random sample
            xi = X_b[random_idx:random_idx+1]
            yi = y[random_idx:random_idx+1]

            # Compute prediction and error for single sample
            prediction = xi.dot(theta)
            error = prediction - yi

            # Compute gradient for single sample
            gradients = xi.T.dot(error)

            # Update weights
            theta = theta - learning_rate * gradients

        # Compute and store cost (MSE) for the entire dataset (for monitoring)
        predictions = X_b.dot(theta)
        total_error = predictions - y
        cost = (1/(2*m)) * np.sum(total_error ** 2)
        cost_history.append(cost)

    return theta, cost_history

theta, cost_history = stochastic_gradient_descent(X, y, learning_rate=0.01, n_epochs=50)
print("Optimized weights:", theta.ravel())
```

```python
plt.plot(range(50), cost_history)
plt.xlabel("Epochs")
plt.ylabel("Cost (MSE)")
plt.title("Cost Reduction Over Epochs (SGD)")
plt.show()
```
